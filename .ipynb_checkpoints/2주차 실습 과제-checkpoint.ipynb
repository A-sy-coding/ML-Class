# KNN 알고리즘

* 지도학습 모델의 한 종류이다.
* 훈련데이터를 학습시킨 후(훈련데이터들은 클래스들을 가지고 있다.) 새로운 데이터를 집어넣어
* 새로운 데이터로부터 가까운 샘플들 k를 선택한 후 과반수 투표를 이용해 예측값을 도출한다.

```python
# knn 알고리즘 
from sklearn.neighbors import KNeighborsClassifier

# knn 객체 만들기
knn = KNeighborsClassifier(n_neighbors=1)

# 모델 학습시키기
knn.fit(X_train,y_train)

# 학습시킨 모델을 이용해 예측하기
y_pred = knn.predict(X_test)
print('테스트 세트에 대한 예측값 : \n',y_pred)  # 각 인덱스의 위치값이 출력된다.
print('예측값을 해당 target name에 매칭 : \n',iris_data.target_names[y_pred]) # 위치값을 집어넣어 해당 target를 출력한다.

# 정확도 평가 --> 2가지 방법
print('테스트 세트의 정확도 : {:.2f}'.format(np.mean(y_test==y_pred)))
print('테스트 세트의 정확도 : {:.2f}'.format(knn.score(X_test,y_test)))

# 오차행렬 구하기
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```

```python
# n_neighbors의 값을 변경할 수 있도록 함수로 구현하기!!

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

def model_k_change(data,k): # data는 데이터 이름 , k는 가까운거리 샘플들 개수 
    
    # train, test 나누기
    X_train,X_test,y_train,y_test = train_test_split(data['data'],data['target'],random_state = 42, test_size = 0.33)
    
    # knn 모델 만들고 학습시키기
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    
    print('n_neighbors : ',k)
    
    # 학습시킨 모델을 이용해 예측하기
    y_pred = knn.predict(X_test)
    #print('테스트 세트에 대한 예측값 : \n',y_pred)  # 각 인덱스의 위치값이 출력된다.
    #print('예측값을 해당 target name에 매칭 : \n',iris_data.target_names[y_pred]) # 위치값을 집어넣어 해당 target를 출력한다.

    # 정확도 평가 --> 2가지 방법
    print('테스트 세트의 정확도 : {:.2f}'.format(np.mean(y_test==y_pred)))
    print('테스트 세트의 정확도 : {:.2f}'.format(knn.score(X_test,y_test)))

    # 오차행렬 구하기
    print('오차행렬 : \n',confusion_matrix(y_pred,y_test))
```

```python
# n_neighbors가 1~10 일때의 정확도를 각각 비교해보기!!
for i in range(1,11):
    model_k_change(iris_data,i)
    
# 정확도가 전부 0.98로 같게 나온다. 
# random_state = 42를 지정했기 때문에 분할된 데이터는 항상 같을 것이고, 다른 부분은 이웃 점들(k)을 얼마로 지정했는가의 차이안데, 
# 정확도가 모두 같다는 것은 코딩이 잘못 되었다는 의미이거나 이웃 점들의 개수에 영향을 받지 않는다는 의미가 된다.

# 만약 코딩이 잘못되었다면 그 이유에 대한 답은 knn의 알고리즘을 좀 더 자세히 들여다 봐야 할 것 같습니다..
```

* 임의의 데이터 만들어보기

```python
# 임의의 분류 데이터셋 만들기 ---> forge
from sklearn.datasets import make_blobs
import numpy as np

def make_forge():
    # centers -> 생성할 클러스터의 수 , n_samples -> 표본 데이터 수 
    # make_bolbs --> 가우시안 정규분포를 이용해 가상 데이터 생성
    # X는 (30,2) 배열 생성   ---> 독립변수
    # y는 (30,1) 배열 생성   ---> 종속변수
    X,y = make_blobs(centers=2, random_state = 4,n_samples = 30)
    
    # y 데이터에서 7,27 인덱스를 0으로 변환
    y[np.array([7,27])] = 0
    
    # X의 길이만큼 1로 채우고 type을 boolean형식으로 변환 (true,false)
    mask = np.ones(len(X),dtype = bool)
    
    # mask에서 0,1,5,26 인덱스를 0(false)으로 변환
    mask[np.array([0,1,5,26])] = 0
    
    # X,y데이터에서 mask가 true인 행들만을 출력한다.
    X,y = X[mask],y[mask]
    
    
    return X,y
```

```python
# 임의의 회귀 데이터셋 만들기 ---> wave


def make_wave(n_samples= 100):
    
    # 시드값 지정
    rnd = np.random.RandomState(42)
    
    # 균일분포 생성
    x = rnd.uniform(-3,3,size=n_samples)
    
    # 정규분포 이용해서 y값 생성
    y_no_noise = np.sin(4*x) + x
    y = (y_no_noise + rnd.normal(size=len(x))) / 2
    
    return x.reshape(1,-1),y
```

* knn 알고리즘 적용 단계
* 1. 데이터를 train,test로 나누기
* 2. knn모델 객체 만들기
* 3. knn모델에 train 데이터 학습시키기
* 4. 교차검증과 같은 모델 평가를 통해 매개변수 지정하기
* 5. 학습된 모델에 test 데이터를 넣어 예측하기

* forge 데이터에 knn 알고리즘 적용

```python
# forge 데이터에 knn 알고리즘 적용
from sklearn.model_selection import train_test_split

X,y = mglearn.datasets.make_forge()

# train, test로 데이터 나누기
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# knn 객체 생성 및 파라미터 설정
clf = KNeighborsClassifier(n_neighbors=3)

# 모델 학습
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

# 교차검증을 사용하여 모델의 정확도 측정하기
print('정확도 : {:.3f} '.format(np.mean(cross_val_score(clf,X_train,y_train,cv=3,scoring='accuracy'))))

# 교차검증을 통한 train데이터 예측
y_train_pred = cross_val_predict(clf,X_train,y_train,cv=3)
# train데이터에서 모델의 오차행렬 구하기
confusion_matrix(y_train_pred,y_train)
```

```python
# test 데이터를 넣어 예측하기
print('테스트 예측 : ',clf.predict(X_test))

# test 데이터 정확도 측정
print('테스트 데이터 정확도 : {:.3f}'.format(clf.score(X_test,y_test)))

# 오차행렬 확인하기
print('오차 행렬 : \n',confusion_matrix(clf.predict(X_test),y_test)) # 1개를 오분류 했다.
```

```python
# 결정경계 확인해보기

# 그래프 크기 설정 및 1행 3열의 서브그래프 설정
fig,axes = plt.subplots(1,3,figsize=(10,3))


# n_neighbors의 값을 1,3,9 순서대로 설정(반복)하여 결정경계 확인
for n_neighbors, ax in zip([1,3,9],axes):
    
    # 객체 생성 및 학습
    clf = KNeighborsClassifier(n_neighbors = n_neighbors).fit(X_train,y_train)
    # 경계선기준으로 색깔 칠하기
    mglearn.plots.plot_2d_separator(clf,X,fill=True,eps=0.5,ax=ax,alpha=.4)
    
    # 클래스들 추가하기
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    
    # 각 축의 그래프 제목을 n_neighbor에 따라 출력하기
    ax.set_title("{}이웃".format(n_neighbors))  # 1이웃, 3이웃, 9이웃
    
    # x축, y축 이름 설정
    ax.set_xlabel('특성0')
    ax.set_ylabel('특성1')
    
# 첫번째 축의 그래프에 범례 추가
axes[0].legend(loc=3)

# 이웃이 1일때는 훈련데이터에 너무 과대적합되어 새로운 데이터에 잘 적용되지 않을 가능성이 높다.
# 과대적합을 해결하기 위해서는 n_neighbor의 개수를 증가시켜야 한다.(단, 너무 많이 증가시키면 과소적합될 가능성이 존재한다.)
```

* 유방암 데이터에 knn 적용

```python
from sklearn.datasets import load_breast_cancer

# 데이터 불러오기
cancer = load_breast_cancer()

# train, test 나누기
# stratify --> 지정한 데이터의 비율을 유지한다.  ---> 데이터를 분할할때 지정한 데이터의 비율을 유지한다.
X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state = 66)
```

```python
# 훈련데이터 정확도와 테스트 데이터 정확도를 저장할 리스트 만들기
training_accuracy = []
test_accuracy = []

# n_neighbors를 1~10까지 바꿔가며 반복하기
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    # 모델 훈련
    clf.fit(X_train,y_train)
    # 훈련 정확도 저장
    training_accuracy.append(clf.score(X_train,y_train))
    # 테스트 정확도 저장
    test_accuracy.append(clf.score(X_test,y_test))
    
# 저장한 데이터들을 이용해 훈련 정확도와 테스트 정확도 시각화 하기
plt.plot(neighbors_settings,training_accuracy,label='훈련 정확도')
plt.plot(neighbors_settings,test_accuracy,label = '테스트 정확도')
plt.xlabel('n_neighbors')
plt.ylabel('정확도')
plt.legend()
```

```python
# n_neoghbors의 값을 1부터 30까지로 늘려서 훈련 정확도와 테스트 정확도 확인해보기

# 훈련데이터 정확도와 테스트 데이터 정확도를 저장할 리스트 만들기
training_accuracy = []
test_accuracy = []

# n_neighbors를 1~10까지 바꿔가며 반복하기
neighbors_settings = range(1,31)

for n_neighbors in neighbors_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    # 모델 훈련
    clf.fit(X_train,y_train)
    # 훈련 정확도 저장
    training_accuracy.append(clf.score(X_train,y_train))
    # 테스트 정확도 저장
    test_accuracy.append(clf.score(X_test,y_test))
    
# 저장한 데이터들을 이용해 훈련 정확도와 테스트 정확도 시각화 하기
plt.plot(neighbors_settings,training_accuracy,label='훈련 정확도')
plt.plot(neighbors_settings,test_accuracy,label = '테스트 정확도')
plt.xlabel('n_neighbors')
plt.ylabel('정확도')
plt.legend()
```

* 훈련 정확도와 테스트 정확도의 그래프를 확인해 보면, 
* 가장 좋은 매개변수로 n_neighbors = 7 정도가 적절할 것 같다고 판단할 수 있다.
* 더 높은 정확도를 얻기 위해서는 데이터 전처리 과정을 추가로 하거나 다른 모델을 사용하는 방법 등이 존재한다.
